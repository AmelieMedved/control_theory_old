# Введение

Задача о стабилизации перевёрнутного маятника является одной из классических учебных задач теории управления. Простота интерпретации параметров задачи и её нелинейность делают её подходящей для демонстрации функционирования регуляторов различных типов. Целью настоящей лабораторной работы является знакомство обучающихся с возможностями и ограничениями различных регуляторов на примере этой классической задачи, а также освоение обучающимися базовых навыков численного исследования систем управления. 

# Объект управления и его модель

В настоящей работе в качестве **объекта управления (_англ._ plant, control object, process)** предполагается механический маятник с осью подвеса, закрепленной на подвижной каретке, которая может свободно перемещаться вдоль некоторой горизонтальной (то есть перпендикулярной к направлению силы тяжести) направляющей рейки. Эта рейка перпендикулярна оси подвеса маятника, так что движение маятника можно считать плоским. В качестве **управления (управляющего воздействия, _англ._ control signal, input signal, control input)** выступает внешняя сила, прикладываемая к каретке параллельно направляющей рейке. Без управления маятник имеет два положения равновесия: устойчивое нижнее и неустойчивое верхнее (перевёрнутое). Задача управления состоит в обеспечении устойчивости верхнего положения равновесия, то есть в обеспечении **целевого режима (_англ._ goal mode)**, при котором положение маятника остаётся в некотором смысле близким в верхнему положению равновесия.

Для построения регулятора можно использовать следующую естественную **физическую модель (_англ._ model)** такого маятника, схематично изображённой на рис.&nbsp;1. Точечный груз массы $M > 0$ закреплён на конце жёсткого невесомого стержня длины $l > 0$. Стержень может свободно вращаться вокруг перпендикулярной к нему оси, проходящей через противоположной к грузу конец стержня. Ось вращения закреплена на точечной каретке массы $m > 0$, которая может перемещаться вдоль оси $x$ декартовой системы координат. К каретке прикладывается переменная внешняя сила, проекция которой на ось $x$ равна $F(t)$, $t$&nbsp;— время. Сила тяжести направлена вдоль отрицательного направления оси $y$ и характеризуется ускорением свободного падения $g \geqslant 0$. В системе присутствует вязкое трение между грузом и окружающей средой, между осью и осевым отверстием и между кареткой и напрявляющей рейкой или окружающей средой, характеризуемое соответственно тремя временными декрементами $\chi_1$, $\chi_2$ и $\chi_3$. **Конфигурацию системы (_англ._ configuration of the system)** в некоторый момент времени $t$ можно полностью охарактеризовать с помощью двух переменных. В качестве таких переменных удобно выбрать безразмерные переменные: нормированную на $l$ абсциссу каретки $\xi(t)$ и ориентированный угол $\varphi(t)$ между положительным направлением оси $y$ и направлением стержня от оси вращения к грузу.

Консервативная динамика такой системы может быть описана функцией Лагранжа $\mathcal L$, которая в данном случае представляет собой разность кинетической $K$ и потенциальной $\Pi$ энергий, $\mathcal L = K - \Pi$. Кинетическая энергия $K = K_1 + K_2$ складывается из кинетической энергии груза $K_1 = Ml^2(\dot\varphi^2 + 2\dot\xi \dot\varphi \cos\varphi  + \dot\xi^2)/2$ и кинетической энергии каретки $K_2 = ml^2\dot\xi^2/2$, потенциальная $\Pi = \Pi_1 + \Pi_2$&nbsp;— из потенциальной энергии груза в поле силы тяжести $\Pi_1 = Mgl\cos\varphi$ и потенциальной энергии взаимодействия каретки с управляющим устройством $\Pi_2 = -Fl\xi$. Таким образом,
$$
\mathcal L = \frac{Ml^2\dot\varphi^2}{2} + Ml^2 \dot\xi \dot\varphi \cos\varphi  + \frac{(m + M)l^2\dot\xi^2}{2} - Mgl \cos\varphi + F(t)l \xi.
$$
Выражение для функции Лагранжа можно упростить до
$$
\tag{1}\label{eq:lagrangian}
\mathcal L = (M + m)l^2 \left(\frac{\alpha\dot\varphi^2 + \dot\xi^2}2 + \alpha\dot\xi\dot\varphi \cos\varphi - \alpha\Omega^2 \cos\varphi + u(t)\xi\right)
$$
с помощью обозначений
$$
\Omega = \sqrt{\frac{g}{l}}, \quad u(t) = \frac{F(t)}{l(M + m)}, \quad \alpha = \frac{M}{M + m},
$$
при этом $0 < \alpha < 1$.

Вязкое трение в системе можно описать с помошью диссипативной функции достаточно общего вида
$$
\tag{2}\label{eq:dissipation}
\mathcal F = (m + M) l^2 \left(\frac{\lambda\alpha\dot\varphi^2 + \mu\dot\xi^2}2 + \nu\alpha\dot\xi\dot\varphi \cos\varphi\right),
$$
где $\lambda \geqslant \nu$, $\mu \geqslant \alpha\nu$, $\nu \geqslant 0$&nbsp;— коэффициенты, характеризующие силы вязкого трения в системе. Эти коэффициенты могут быть выражены через $\chi_{1, 2, 3}$: $\lambda = (\chi_1 + \chi_2)/\alpha$, $\mu = \chi_1 + \chi_3$ и $\nu = \chi_1/\alpha$. В данном случае диссипативная функция&nbsp;[$(2)$](#mjx-eqn-eqdissipation) имеет смысл половинной мощности потерь (энергии, теряемой системой в единицу времени вследствие диссипации) и представляет собой всего-навсего линейную комбинацию квадратов скоростей груза, каретки и угловой скорости вращения маятника вокруг оси подвеса. При этом условно можно считать, что величина $\chi_1 = \alpha\nu$ характеризует силу вязкого трения груза об окружающую среду (например, воздух): модуль этой силы, действующей на груз, равен $\nu Ml \sqrt{\dot\varphi^2 + \dot\xi^2 + 2\dot\varphi\dot\xi \cos\varphi}$. В свою очередь, $\chi_2 = \alpha(\lambda - \nu)$ характеризует силу вязкого трения между осью маятника и осевым отверстием: модуль момента этой силы равен $(\lambda - \nu) Ml^2 |\dot\varphi|$. Величина $\chi_3 = \mu - \alpha\nu$ характеризует силу вязкого трения каретки об окружающую среду и направляющую рейку: её модуль равен $(\mu - \alpha\nu) (M + m) l |\dot\xi|$.

Уравнения Лагранжа с учётом диссипации имеют вид
$$
\tag{3}\label{eq:eulerlagrangephi}
\frac{\partial \mathcal L}{\partial \varphi} - \frac{d}{dt} \frac{\partial \mathcal L}{\partial \dot\varphi} = \frac{\partial \mathcal F}{\partial \dot\varphi},
$$
$$
\tag{4}\label{eq:eulerlagrangexi}
\frac{\partial \mathcal L}{\partial \xi} - \frac{d}{dt} \frac{\partial \mathcal L}{\partial \dot\xi} = \frac{\partial \mathcal F}{\partial \dot\xi}.
$$
Подставляя уравнения&nbsp;[$(1)$](#mjx-eqn-eqlagrangian) и [$(2)$](#mjx-eqn-eqdissipation) в [$(3)$](#mjx-eqn-eqeulerlagrangephi) и [$(4)$](#mjx-eqn-eqeulerlagrangexi)  и упрощая получившиеся выражения, получаем
$$
\tag{5}\label{eq:odephi}
\ddot\varphi + \lambda\dot\varphi + (\ddot\xi + \nu\dot\xi) \cos\varphi - \Omega^2 \sin\varphi =  0,
$$
$$
\tag{6}\label{eq:odexi}
\ddot\xi + \mu\dot\xi + \alpha(\ddot\varphi + \nu\dot\varphi) \cos\varphi - \alpha\dot\varphi^2 \sin\varphi - u = 0.
$$
Получившаяся система уравнений полностью описывает используемую **модель объекта управления (динамическую модель системы, _англ._ plant model, dynamical model of the system)**. В этой системе $\varphi$ и $\xi$&nbsp;— **выходные переменные (_англ._ outputs)**, $u$&nbsp;— **управление (_англ._ control input)** и $\Omega$, $\alpha$, $\lambda$, $\mu$, $\nu$&nbsp;— **параметры системы (_англ._ parameters of the system)**. Напомним, что данная система может представлять модель маятника с осью на подвижной каретке, только если параметры удовлетворяют следующим неравенствам: $0 < \alpha < 1$, $\lambda \geqslant \nu$, $\mu \geqslant \alpha\nu$, $\nu \geqslant 0$. Также без ограничения общности можно считать, что $\Omega \geqslant 0$.

Систему&nbsp;[$(5)$](#mjx-eqn-eqodephi) и [$(6)$](#mjx-eqn-eqodexi) можно переписать для векторной переменной
$$
\mathbf x \equiv \begin{bmatrix}
x_0 \\
x_1 \\
x_2 \\
x_3
\end{bmatrix} \equiv \begin{bmatrix}
\varphi \\
\dot\varphi \\
\xi \\
\dot\xi
\end{bmatrix}
$$
как
$$
\tag{7}\label{eq:vectorode}
\dot{\mathbf x} = \frac{1}{1 - \alpha \cos^2 x_0} \begin{bmatrix}
\left(1 - \alpha \cos^2 x_0\right) x_1 \\
\left(\Omega^2 - \alpha x_1^2 \cos x_0\right) \sin x_0 - \left(\lambda - \alpha\nu \cos^2 x_0\right) x_1 + (\mu - \nu) \cos x_0 \, x_3 - \cos x_0 \, u\\
\left(1 - \alpha \cos^2 x_0\right) x_3 \\
-\alpha \left(\Omega^2 \cos x_0 - x_1^2\right) \sin x_0 + \alpha (\lambda - \nu) \cos x_0 \, x_1 - \left(\mu - \alpha\nu \cos^2 x_0\right) x_3 + u
\end{bmatrix}.
$$
В таком виде система будет использоваться далее при численном интегрировании.

Система с нулевым управлением $u \equiv 0$, является автономной системой, заданной на четырёхмерном **фазовом пространстве (_англ._ phase space)** $(\varphi, \dot\varphi, \xi, \dot\xi)$. В теории управления фазовое пространство часто также называют **пространством состояний (_англ._ state space)**. Разомкнутая система симметрична относительно замены $\varphi \rightarrow \varphi + 2\pi n$, $n$&nbsp;— целое. Эта симметрия связана с тем, что углы $\varphi$, отличающиеся на величину, кратную $2\pi$, в реальной системе отвечают одному и тому же положению маятника. Благодаря этой симметрии всё фазовое пространство (пространство состояний) можно редуцировать до четырёхмерного фазового цилиндра, на котором $-\pi < \varphi \leqslant \pi$. На этом фазовом цилиндре разомкнутая система имеет бесконечное множество состояний равновесия, образующее две прямые: состояния равновесия на прямой $\varphi = 0$, $\dot \varphi = 0$, $\dot \xi = 0$ (с произвольным постоянным значением $\xi$) соответствуют покоящемуся маятнику с грузом в верхней точке (верхнем положении), а состояния на прямой $\varphi = \pi$, $\dot \varphi = 0$, $\dot \xi = 0$ отвечают покоящемуся маятнику с грузом в нижнем положении. Линеаризовав систему&nbsp;[$(7)$](#mjx-eqn-eqvectorode) вблизи одного из верхних равновесных состояний
$$
\mathbf x = \mathbf x_0 = \begin{bmatrix}
0 \\
0 \\
\xi_0 \\
0
\end{bmatrix}
$$
с произвольным $\xi_0$, получим линейную систему обыкновенных дифференциальных уравнений
$$\dot{\mathbf{\delta x}} = \mathbf A\mathbf{\delta x},$$
где $\mathbf{\delta x} = \mathbf x - \mathbf x_0$&nbsp;— малое отклонение от равновесного состояния и $\mathbf A$&nbsp;— матрица системы,
$$\mathbf A = \frac{1}{1 - \alpha} \begin{bmatrix}
0 & 1 - \alpha & 0 & 0 \\
\Omega^2 & -\lambda + \alpha\nu & 0 & \mu - \nu \\
0 & 0 & 0 & 1 - \alpha \\
-\alpha\Omega^2 & \alpha (\lambda - \nu) & 0 & -\mu + \alpha\nu
\end{bmatrix}.$$
Характеристический полином мтарицы $\mathbf A$ имеет вид
$$
P(p) = p^4 + \frac{\lambda  + \mu - 2\alpha  \nu}{1 - \alpha} p^3 - \left(\frac{\Omega^2 + \alpha\nu ^2 - \lambda\mu}{1 - \alpha} - \alpha \frac{(\mu - \nu)^2}{(1 - \alpha)^2}\right) p^2 - \frac{\mu\Omega^2}{1 - \alpha} p.
$$
Один из корней этого полинома равен $0$, а три других удовлетворяют кубическому уравнению
$$
p^3 + \frac{\lambda  + \mu - 2\alpha  \nu}{1 - \alpha} p^2 - \left(\frac{\Omega^2 + \alpha\nu ^2 - \lambda\mu}{1 - \alpha} - \alpha \frac{(\mu - \nu)^2}{(1 - \alpha)^2}\right) p - \frac{\mu\Omega^2}{1 - \alpha} = 0.
$$
Свободный член в этом уравнении отрицателен, что означает, что у этого уравнения есть по крайней мере один действительный положительный корень и состояние равновесия является неустойчивым. Нетрудно показать, что нижние положения равновесия при это оказываются нейтральными. Малые отклонения от верхнего положения равновесия со временем нарастают и груз опускается вниз, стремясь к одному из нижних положений равновесия, что иллюстрирует видео&nbsp;1. Таким образом, объект управления является неустойчивым.

Для стабилизации верхнего положения равновесия можно применить один из наиболее общих принципов построения систем управления&nbsp;— принцип **обратной связи (_англ._ feedback, feedback signal)**, который предполагает формирование управляющего воздействия на основе измеряемых или оцениваемых отклонений выходных переменных от целевых значений. При реализации принципа обратной связи **измерительное устройство (_англ._ sensor)** определяет некоторые данные о состоянии системы, которые затем поступают на вход **управляющего устройства (_англ._ controller)**, вырабатывающего управляющее воздействие, реализуемое с помощью исполнительного устройства **исполнительного устройства (_англ._ actuator)**. Объект управления, измерительное, управляющее и исполнительное устройства вместе формируют **замкнутую систему управления с обратной связью (_англ._ closed-loop feedback control system)**.

В данной работе будет рассматриваться наиболее простая ситуация, когда измерительное, управляющее и исполнительные устройства позволяют реализовать **обратную связь по состоянию (_англ._ state feedback)** с управляющим воздействием, определяемым текущим состоянием маятника $(\varphi, \dot\varphi, \xi, \dot\xi)$. Это, в частности, предполагает мгновенную реакцию этих устройств, без временных лагов (задержек). На практике это означает, что характерные времена, требуемые для измерения состояния и формирования управляющего воздействия, значительно меньше, чем собственные времена объекта управления, определяемые параметрами $\Omega$, $\lambda$, $\mu$, $\nu$ и $\alpha$. Таким образом, **стратегия управления (_англ._ control strategy, control policy)**&nbsp;— функция, определяющая управляющее воздействие&nbsp;— в данной работе имеет вид
$$
\tag{8}\label{eq:control}
u = U(\varphi, \dot\varphi, \xi, \dot\xi).
$$
При таком выборе стратегии управления математическая модель замкнутой системы с обратной связью, также иногда называемой **замкнутой системой (_англ._ closed-loop system)**, получается при подстановке управления&nbsp;[$(8)$](#mjx-eqn-eqcontrol) в уравнения&nbsp;[$(5)$](#mjx-eqn-eqodephi) и [$(6)$](#mjx-eqn-eqodexi). Эта модель является автономной системой, заданной на четырёхмерном фазовом пространстве, также как и исходная **разомкнутая система (_англ._ open-loop system)** с нулевым управлением.

Так как обычно не важно, сколько оборотов совершит маятник, прежде чем придёт к вертикальному положению, можно говорить о том, что заданный целевой режим регулирования не различает углы, отличающиеся на величину, кратную $2\pi$. Соответственно при построении регулятора также не имеет смысла различать такие углы и следует выбирать стратегии управления $U(\varphi, \dot\varphi, \xi, \dot\xi)$, периодические по $\varphi$ с периодом $2\pi$, $U(\varphi + 2\pi, \dot\varphi, \xi, \dot\xi) = U(\varphi, \dot\varphi, \xi, \dot\xi)$. Такие периодические стратегии управления не разрушают симметрии исходной системы относительно замены $\varphi \rightarrow \varphi + 2\pi$, и эта симметрия сохраняется и в замкнутой системе, а в качестве пространства состояний можно рассматривать четырёхмерный фазовый цилиндр с $-\pi < \varphi \leqslant \pi$ и задавать $U$ только для таких $\varphi$. Если однако $U$ не является периодической функцией $\varphi$, то симметрия теряется и система становится чувствительной к количеству совершённых оборотов маятника. Конкретные стратегии управления, рассматриваемые в данной работе: линейная и рэлейная (и их модификации) приведены в соответствующих блоках заданий.

Если функция $U$ дифференцируема в окрестности состояния равновесия $(\varphi, \dot\varphi, \xi, \dot\xi) = (0, 0, 0, 0)$, то линеаризованная система в стандартной форме записывается как
$$\dot{\mathbf x} = \mathbf A\mathbf x + \mathbf B u,$$
где
$$\mathbf B = \frac{1}{1 - \alpha} \begin{bmatrix}
0 \\
-1 \\
0 \\
1
\end{bmatrix}.
$$
При линейном управлении&nbsp;[$(7)$](#mjx-eqn-eqcontrol) можно записать $u = \mathbf k^T \mathbf x$, где
$$\mathbf k = \begin{bmatrix}
k_0 \\
k_1 \\
k_2 \\
k_3
\end{bmatrix},$$
где $k_0 = \frac{\partial U(0, 0, 0, 0)}{\partial \varphi}$, $k_1 = \frac{\partial U(0, 0, 0, 0)}{\partial \dot\varphi}$, $k_2 = \frac{\partial U(0, 0, 0, 0)}{\partial \xi}$ и $k_3 = \frac{\partial U(0, 0, 0, 0)}{\partial \dot\xi}$&nbsp;— постоянные коэффициенты.

Теория управления&nbsp;— одна из многих областей, где эта связка также используется всё чаще, а знакомство со связкой Jupyter Notebook и Python имеет приоритетную необходимость для студентов, специализирующихся в теории управления.

## Jupyter Notebook

Сервер Jupyter Notebook предоставляет интерактивную оболочку для различных интерпретируемых языков программирования, включая язык Python. Клиенты могут подключаться к серверу посредством веб-интерфейса с помощью приложения-браузера. Отображаемая в браузере страница блокнота (notebook) представляет собой последовательность ячеек (cells), которые могут содержать 1) форматированный текст или заголовки на языке разметки Markdown (Markdown cells); 2) команды (код) интерпретируемого языка программирования и результаты его выполнения вместе с предупреждениями и ошибками выполнения (code cells) и 3) неформатированный текст (raw NBConvert cells). При работе с интерактивным блокнотом можно изменять содержимое ячеек и запускать код отдельных ячеек для выполнения в интерпретаторе (ядре) на стороне сервера. Форматированный текст может содержать математические формулы и гиперссылки. Результаты выполнения кода помимо символьной информации могут содержать произвольный код на языке разметки HTML, в частности код для рисунков, видео, интерактивных элементов и виджетов, которые отрисовываются на стороне клиента. По сути блокнот Jupyter Notebook (и иные интерфейсы блокнота, например Mathematica Notebook) представляют собой реализацию концепции грамотного программирования (literate programming), предложенной Дональдом Кнутом. Справка по интерфейсу блокнота (в частности, сочетания клавиш клавиатуры для запуска кода на выполнение, переключения между типами ячеек, включения и отключения режима правки ячеек) доступна из меню `Help` в строке меню на странице блокнота. 

В [прилагаемом блокноте](student_part1.ipynb) содержатся незаполненные или недозаполненные клетки для форматированного текста и кода, которые предлагается заполнить обучающимся в соответствии с формулированными перед этими клетками заданиями.
